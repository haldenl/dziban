````python
import altair as alt
alt.renderers.enable('notebook')
from vega_datasets import data
````

````python
cars = data.cars()
movies = data.movies()
````

# Example 1

A user wants to visualize the cars dataset. They start with the question 'I wonder how miles per gallon relates to horsepower?'

They could formulate this query a such:

```
dziban.Chart(cars).encode(['Horsepower', 'Miles_per_Gallon'])
```

With the following result

````python
alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon'
)
````

Here, they spot a trend.

````python
alt.Chart(movies).mark_point().encode(
    x=alt.X('IMDB_Rating:Q'),
    y=alt.Y('Rotten_Tomatoes_Rating:Q')
)
````

But now they want to see this normalized. The specification would be as follows.

````python
alt.Chart(movies).mark_bar().encode(
    x=alt.X('IMDB_Rating:Q', bin=True),
    y=alt.Y('count()', stack='normalize')
)
````
