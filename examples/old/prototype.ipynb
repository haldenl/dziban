```python
from dziban.chart import Chart
from vega_datasets import data
```

```python
chart = Chart(data.cars())
```

```python
chart['Horsepower'].aggregate('mean').channel('x')
```

```python
chart.see('Horsepower', 'Miles_per_Gallon', name='first')
```

```python
chart['Horsepower'].scale('log')
chart.see('Horsepower', 'Miles_per_Gallon', anchor='first')
```
